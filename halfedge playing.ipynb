```python
import open3d as o3d
import numpy as np
import trimesh
from bns_utils import compute_onering_data
from mesh_processing import get_V_he, optimise_V_he


```

```python
coarse_mesh_o3d = o3d.io.read_triangle_mesh('data/surfaces/two_triangles.obj')

        
he_mesh = o3d.geometry.HalfEdgeTriangleMesh.create_from_triangle_mesh(coarse_mesh_o3d)
V_he = get_V_he(he_mesh)
V_he = optimise_V_he(he_mesh, V_he)
print(V_he)

```

```python

onerings = compute_onering_data(he_mesh, V_he=V_he)
print( [( onering['V_indices'], onering['triangles'], onering['valence'], onering['is_boundary']) for onering in onerings] )
```
