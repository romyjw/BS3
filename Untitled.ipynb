```python
from colourmappings import *



import matplotlib.pyplot as plt
import matplotlib.colorbar as colorbar
import matplotlib.colors as mcolors
import numpy as np

# 1. Setup the figure and axes
fig, ax = plt.subplots(figsize=(8, 2))
fig.subplots_adjust(bottom=0.5)

# 2. Define your normalization (adjust vmin/vmax to your data range)
norm = mcolors.Normalize(vmin=-1.0, vmax=1.0)

# 3. Create the colorbar using your custom mapping function 'mapping12'
# If mapping12 is a Colormap object:
cb = colorbar.ColorbarBase(ax, cmap=mapping12,
                           norm=norm,
                           orientation='horizontal')

# 4. Styling for Seismic aesthetics
cb.set_label('Amplitude / Velocity')
ax.set_title('Seismic Data Mapping (mapping12)')

plt.show()
```
