```python
import torch
import torch.nn as nn
import torch.optim as optim

# Define a simple MLP with customizable layer sizes
class MLP(nn.Module):
    def __init__(self, layer_sizes):
        super(MLP, self).__init__()
        layers = []
        for i in range(len(layer_sizes) - 1):
            layers.append(nn.Linear(layer_sizes[i], layer_sizes[i + 1]))
            if i < len(layer_sizes) - 2:
                layers.append(nn.Tanh())  # activation between hidden layers
        self.model = nn.Sequential(*layers)

    def forward(self, x):
        return self.model(x)




```

```python
def disp_modifier_orthog(bs_points, bs_normals, model, scale=0.1):
    disp = model(bs_points)
    return bs_points + disp*scale*bs_normals

def disp_modifier_general(bs_points, bs_normals, model, scale=0.01):
    disp = model(bs_points)
    return bs_points + disp*scale
```

```python
layer_sizes = [3, 64, 3]

layer_sizes = [3, 64, 64, 1]
disp_mlp = MLP(layer_sizes).to(device)





```

```python
import torch

# ---- Precompute samples ----
training_sample_dict = bns.compute_samples(num_samples=10000, avoid_corners=True)
x_all = training_sample_dict['uv'].to(device)
x_all.requires_grad_(True)

bns.disp_modifier = None

# ---- Compute bs_points and bs_normals WITH gradients ----
bs_points_all = bns(x_all, test_flag='poly')
bs_normals_all = diffmod.compute_normals(out=bs_points_all, wrt=x_all)

# Detach after both have grad-fns
bs_points_all = bs_points_all.detach()
bs_normals_all = bs_normals_all.detach()

```

```python
# ---- Training loop ----
num_epochs = 100000
batch_size = 1024
num_samples = x_all.shape[0]

disp_mlp.to('mps')
optimizer = optim.Adam(disp_mlp.parameters(), lr=0.001)



for epoch in range(num_epochs):
    perm = torch.randperm(num_samples, device=device)
    
    for i in range(0, num_samples, batch_size):
        idx = perm[i:i + batch_size]

        # Re-enable gradient tracking per batch
        bs_points = bs_points_all[:, idx, :].clone().requires_grad_(True)
        bs_normals = bs_normals_all[:, idx, :].clone().requires_grad_(True)

        optimizer.zero_grad()

        # Apply displacement modifier (your learned model)
        new_points = disp_modifier_general(bs_points, bs_normals, model=disp_mlp)

        # Compute loss (e.g., SDF)
        loss = sdf(new_points, shape_id, squared=True).mean()

        # Backpropagate and update
        loss.backward()
        optimizer.step()

    # ---- Logging ----
    if epoch == 0:
        with torch.no_grad():
            baseline_loss = sdf(bs_points_all, shape_id, squared=True).mean()
        print(f"Epoch {epoch}: baseline loss = {baseline_loss.item():.10f}")

    if epoch % 100 == 0:
        print(f"Epoch {epoch}: loss = {loss.item():.10f}")
```
