```python
import math

def generate_circle_obj(filename="circle.obj", radius=1.0, num_vertices=200):
    """
    Generate a .obj file representing a circle centered at the origin.
    
    Parameters:
        filename (str): Output .obj filename
        radius (float): Circle radius
        num_vertices (int): Number of vertices around the circle
    """
    
    with open(filename, 'w') as f:
        f.write("# Circle OBJ file\n")
        f.write(f"# {num_vertices} vertices, radius = {radius}\n\n")

        # Write vertices
        for i in range(num_vertices):
            angle = 2 * math.pi * i / num_vertices
            x = radius * math.cos(angle)
            y = radius * math.sin(angle)
            z = 0.0
            f.write(f"v {x:.6f} {y:.6f} {z:.6f}\n")
        
        f.write("\n# Edges (lines)\n")

        # Write edges (as line segments)
        # Each vertex connects to the next, and the last connects back to the first
        for i in range(1, num_vertices):
            f.write(f"l {i} {i+1}\n")
        f.write(f"l {num_vertices} 1\n")  # Close the circle

    print(f"✅ .obj file successfully written to '{filename}'")

# Example usage
if __name__ == "__main__":
    generate_circle_obj("coarse_circle.obj", radius=1.0, num_vertices=5)
```
