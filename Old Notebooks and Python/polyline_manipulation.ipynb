```python
import trimesh
import numpy as np

mesh = trimesh.load('data/curves/shark_bdry.obj')
verts = mesh.vertices  # Already ordered polyline

# Sample 30 points without repeating the first/last
idx = np.linspace(0, len(verts) - 1, 30, endpoint=False, dtype=int)
new_verts = verts[idx]

write_polyline_to_obj(V=new_verts, filename='data/curves/shark30.obj')

```
