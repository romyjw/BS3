```python
import os
import shutil

shape_name = 'sphere2'
original_cwd = os.getcwd()

# Define paths
patches_path = os.path.join('patches', shape_name)
slim_path = os.path.abspath('../Scalable-Locally-Injective-Mappings')
temp_data_path = os.path.join(slim_path, 'temp_data')
temp_results_path = os.path.join(slim_path, 'temp_results')

# Copy shape directory into temp_data
shutil.rmtree(temp_data_path, ignore_errors=True)  # Clear existing temp_data if needed
shutil.copytree(patches_path, temp_data_path)

# Change to SLIM directory
os.chdir(slim_path)

# Create temp_results if not exist
os.makedirs(temp_results_path, exist_ok=True)

# Process all files in temp_data
for file_name in os.listdir('temp_data'):
    file_path = os.path.join('temp_data', file_name)
    if os.path.isfile(file_path):
        output_file = os.path.join('temp_results', file_name[:-4] + '_param.obj')
        os.system(f'binaries/macos-x86_64/ReweightedARAP {file_path} {output_file}')

# Move results back to patches/shape_name
for result_file in os.listdir(temp_results_path):
    src = os.path.join(temp_results_path, result_file)
    dst = os.path.join(original_cwd, 'patches', shape_name, result_file)
    shutil.move(src, dst)

# Return to original directory
os.chdir(original_cwd)
```
