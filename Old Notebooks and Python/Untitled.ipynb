```python
import open3d as o3d

print("Open3D version:", o3d.__version__)

# Create a simple test mesh (sphere)
mesh = o3d.geometry.TriangleMesh.create_sphere(radius=1.0)
mesh.compute_vertex_normals()

# Create a material
material = o3d.visualization.rendering.MaterialRecord()
material.shader = "defaultLit"   # requires new renderer
material.base_color = [0.2, 0.7, 0.9, 1.0]  # RGBA
material.base_roughness = 0.3
material.base_metallic = 0.8

# Try the new visualizer
try:
    vis = o3d.visualization.O3DVisualizer("New O3DVisualizer Test", 800, 600)
    vis.add_geometry("sphere", mesh, material)
    vis.reset_camera_to_default()
    vis.show(True)
    print("✅ Success: New O3DVisualizer works!")
except Exception as e:
    print("❌ Could not start O3DVisualizer:", e)
```

```python
import open3d as o3d

# Sphere geometry
mesh = o3d.geometry.TriangleMesh.create_sphere(radius=1.0)
mesh.compute_vertex_normals()

# Diffuse-only material
mat = o3d.visualization.rendering.MaterialRecord()
mat.shader = "defaultLit"           # lit shader (enables diffuse lighting)
mat.base_color = (0.8, 0.2, 0.2, 1.0)  # red
mat.base_metallic = 0.0             # non-metallic (no specular reflection)
mat.base_roughness = 1.0            # very rough → kills specular highlights

# Show in interactive window
o3d.visualization.draw([{"name": "sphere", "geometry": mesh, "material": mat}])
```
