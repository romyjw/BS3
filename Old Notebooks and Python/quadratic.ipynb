```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the (x, y) grid
x = np.linspace(-2, 2, 50)
y = np.linspace(-2, 2, 50)
X, Y = np.meshgrid(x, y)

# Define coefficients for each component of the vector-valued function
# Each component is a quadratic: a + bx + cy + dx^2 + exy + fy^2

# Component 1 (U, X direction)
a1, b1, c1, d1, e1, f1 = 2, 1, 0, 0, 0, 0
# Component 2 (V, Y direction)
a2, b2, c2, d2, e2, f2 = 3, 2, 1, -0.3, 0.4, 0
# Component 3 (W, Z direction)
a3, b3, c3, d3, e3, f3 = 0, 0, 0, 0.2, 0.1, 0.3

# Evaluate the parametric surface
U = a1 + b1*X + c1*Y + d1*X**2 + e1*X*Y + f1*Y**2
V = a2 + b2*X + c2*Y + d2*X**2 + e2*X*Y + f2*Y**2
W = a3 + b3*X + c3*Y + d3*X**2 + e3*X*Y + f3*Y**2

# Plot the surface
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(U, V, W, facecolors=plt.cm.viridis((W - W.min()) / (W.max() - W.min())),
                rstride=1, cstride=1, antialiased=True, linewidth=0)

ax.set_title("Parametric Surface: Quadratic Mapping from ℝ² to ℝ³")
ax.set_xlabel("U (X component)")
ax.set_ylabel("V (Y component)")
ax.set_zlabel("W (Z component)")

plt.tight_layout()
plt.show()
```
