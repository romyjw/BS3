```python
def parse_off_file(input_file, output_file):
    with open(input_file, 'r') as infile:
        lines = infile.readlines()
    
    if lines[0].strip() == "OFF":
        lines = lines[1:]
    
    header = list(map(int, lines[0].strip().split()))
    if len(header) == 3:
        num_vertices, num_faces, _ = header  # Ignore the number of edges
    else:
        raise ValueError("Invalid OFF file header")
    
    vertices = lines[1:num_vertices + 1]
    faces = lines[num_vertices + 1:num_vertices + 1 + num_faces]
    
    cleaned_faces = []
    for face in faces:
        indices = list(map(int, face.strip().split()))
        cleaned_faces.append(" ".join(map(str, indices[1:])))  # Strip number of vertices per face
    
    with open(output_file, 'w') as outfile:
        outfile.write(f"{num_vertices} {num_faces}\n")
        outfile.writelines(vertices)
        outfile.writelines(cleaned_face + "\n" for cleaned_face in cleaned_faces)
    
    print(f"Processed file saved as {output_file}")

# Example usage
# parse_off_file("input.off", "output.off")
```

```python
import trimesh
import open3d as o3d

import numpy as np
import open3d as o3d
import copy
import matplotlib.pyplot as plt

def edge_split(he_mesh, V_he, edge_id, vtx_position=None, test_noise=False):
    """
    Splits an edge in a half-edge mesh into two edges by inserting a new vertex.

    :param he_mesh: The half-edge mesh (o3d.geometry.HalfEdgeTriangleMesh)
    :param edge_id: The index of the edge to be split
    :param new_vtx_id: The index of the new vertex
    :return: Updated half-edge mesh
    """
    '''
    for he in he_mesh.half_edges:
        print(he)
    '''
    
    #print(he_mesh.half_edges)

    
    #print(len(he_mesh.triangles))

    old_he_mesh = copy.deepcopy(he_mesh)

    V = np.asarray(he_mesh.vertices)

    next_vtx_id = len(he_mesh.vertices)
    next_triangle_id = len(he_mesh.triangles)

    # Get original half-edges
    he = he_mesh.half_edges[edge_id]
    if he.twin is None:
        raise ValueError(f"Half-edge {edge_id} has no twin, cannot split.")
    he_twin = he_mesh.half_edges[he.twin]

    ################### vertices ##############
    v1 = he.vertex_indices[0]
    v2 = he.vertex_indices[1]
    v3 = he_mesh.half_edges[ he.next].vertex_indices[1]
    v4 = he_mesh.half_edges[ he_twin.next].vertex_indices[1]

    if vtx_position is None:
        vtx_position = (V[v1,:]+V[v2,:])/2
    if test_noise == True:
        vtx_position += 0.1*np.random.randn(3)

    v5 = len(he_mesh.vertices)
        
    he_mesh.vertices.append(vtx_position) #add midpoint vertex
    
    #################################################


    ###################
                           
    # Assign new triangles
    print('halfedge:',he)
    
    he_mesh.triangles[he.triangle_index] = [next_vtx_id, v2, v3] #update old triangle
    he_mesh.triangles[he_twin.triangle_index] = [next_vtx_id, v4, v2] # update old triangle
    
    
    he_mesh.triangles.append ( [next_vtx_id, v1, v4] ) # new triangle
    he_mesh.triangles.append ( [next_vtx_id, v3, v1] ) # new triangle

    
    
    t1 = he.triangle_index
    t2 = he_twin.triangle_index
    t3 = next_triangle_id
    t4 = next_triangle_id + 1

    #print('triangle ind', t1,t2,t3,t4)

    #############
    

    # Create new half-edges (8 in total)
    new_hes = [o3d.geometry.HalfEdge() for _ in range(8)]

    # Assign new vertex indices
    #print('nxt vtx id', next_vtx_id, v2)
    new_hes[0].vertex_indices = [next_vtx_id, v2]
    new_hes[1].vertex_indices = [v2, next_vtx_id]
    new_hes[2].vertex_indices = [next_vtx_id, v4]
    new_hes[3].vertex_indices = [v4, next_vtx_id]
    
    new_hes[4].vertex_indices = [next_vtx_id, v1]
    new_hes[5].vertex_indices = [v1, next_vtx_id]
    new_hes[6].vertex_indices = [next_vtx_id, v3]
    new_hes[7].vertex_indices = [v3, next_vtx_id]

    next_he_id = len(he_mesh.half_edges)
    
    h0  = edge_id
    h1 = he.twin
    h2,h3,h4,h5,h6,h7 = [next_he_id + i for i in range(6) ]

    a = he_mesh.half_edges[ he_twin.next ].next
    b = he.next
    c = he_mesh.half_edges[ he.next ].next
    d = he_twin.next

    
    # Update twin relationships
    new_hes[0].twin = h1
    new_hes[1].twin = h0

    new_hes[2].twin = h3
    new_hes[3].twin = h2

    new_hes[4].twin = h5
    new_hes[5].twin = h4

    new_hes[6].twin = h7
    new_hes[7].twin = h6

    # Update next half-edge pointers
    new_hes[0].next = b
    new_hes[1].next = h2
    new_hes[2].next = a
    new_hes[3].next = h4

    new_hes[4].next = d
    new_hes[5].next = h6
    new_hes[6].next = c
    new_hes[7].next = h0

    # triangle indices
    new_hes[0].triangle_index = t1
    new_hes[1].triangle_index = t2
    new_hes[2].triangle_index = t2
    new_hes[3].triangle_index = t3
    new_hes[4].triangle_index = t3
    new_hes[5].triangle_index = t4
    new_hes[6].triangle_index = t4
    new_hes[7].triangle_index = t1
    


    # Replace old half-edges in the mesh
    
    he_mesh.half_edges[edge_id].vertex_indices = new_hes[0].vertex_indices
    he_mesh.half_edges[edge_id].twin = new_hes[0].twin
    he_mesh.half_edges[edge_id].next = new_hes[0].next
    he_mesh.half_edges[edge_id].triangle_index = new_hes[0].triangle_index
    
    he_mesh.half_edges[he.twin].vertex_indices = new_hes[1].vertex_indices
    he_mesh.half_edges[he.twin].twin = new_hes[1].twin
    he_mesh.half_edges[he.twin].next = new_hes[1].next
    he_mesh.half_edges[he.twin].triangle_index = new_hes[1].triangle_index


    

    he_mesh.half_edges[a].next = h1
    he_mesh.half_edges[a].triangle_index = t2

    he_mesh.half_edges[b].next = h7
    he_mesh.half_edges[b].triangle_index = t1

    he_mesh.half_edges[c].next = h5
    he_mesh.half_edges[c].triangle_index = t4

    he_mesh.half_edges[d].next = h3
    he_mesh.half_edges[d].triangle_index = t3
    
    # Append new half-edges
    temp = he_mesh.half_edges
    temp.extend(new_hes[2:])
    he_mesh.half_edges = temp


    # update V_he
    V_he.append(h0)
    V_he[v1] = h5
    V_he[v2] = h1
    V_he[v3] = h7
    V_he[v4] = h3



    '''
    for i in range(len(he_mesh.half_edges)):
        print('he ',i)
        print(he_mesh.half_edges[i])
        try:
            print(old_he_mesh.half_edges[i])
        except:
            pass
        print()
        print()
    '''

            
    return he_mesh, V_he
    
'''
def ordered_onering_halfedges(he_mesh, V_he, vtx_id):
    first_he = V_he[vtx_id]
    cur_he = first_he
    
    vertices = []
    halfedges = []
    new_he=True
    while new_he:
        vertices.append(cur_he.vertex_indices[1])
        halfedges.append(cur_he.)
        cur_he = he_mesh.half_edges[ he_mesh.half_edges[ cur_he ].twin ].next
        if cur_he == first_he:
            new_he=False
            
    return vertices, halfedges
'''    
def point_on_segment(p, a, b, tol=1e-8):
    ab = b - a
    ap = p - a

    # Check if colinear 
    cross = np.cross(ab, ap)
    if (cross**2).sum() > tol:
        return False

    # Check if p is between a and b using dot product
    dot = np.dot(ap, ab)
    if dot < 0 or dot > np.dot(ab, ab):
        return False

    return True


def dist(p, a, b):
    ab = b - a
    ap = p - a
    proj_length = np.dot(ap, ab) / np.dot(ab, ab)  # Projection scalar
    proj_point = a + proj_length * ab  # Projection point on line
    return np.linalg.norm(p - proj_point)  # Euclidean distance

    


    
def ordered_onering_halfedges(he_mesh, V_he, vtx_id):
    first_he = V_he[vtx_id]
    cur_he = first_he

    #print(cur_he)
    
    ring_halfedges = []
    new_he=True
    while new_he:
        
        ring_halfedges.append(he_mesh.half_edges[ cur_he ].next)
        cur_he = he_mesh.half_edges[ he_mesh.half_edges[ cur_he ].twin ].next
        if cur_he == first_he:
            new_he=False
            
    return ring_halfedges[::-1]

def cut_mesh_along_geodesic(mesh_filepath, geodesic_filepath, first_vtx_id, last_vtx_id, debug_plot=False):

    print('Preparing mesh')
    
    mesh_tm = trimesh.load(mesh_filepath)
    mesh_o3d= o3d.geometry.TriangleMesh()
    mesh_o3d.vertices=o3d.utility.Vector3dVector( mesh_tm.vertices )
    mesh_o3d.triangles=o3d.utility.Vector3iVector( mesh_tm.faces)

    he_mesh = o3d.geometry.HalfEdgeTriangleMesh.create_from_triangle_mesh(mesh_o3d)


    V_he = [-1] * len(he_mesh.vertices)  # More efficient than list comprehension

    for i, he in enumerate(he_mesh.half_edges):
        if i % 10000 == 0:
            print(i, len(he_mesh.half_edges))
    
        V_he[he.vertex_indices[0]] = i  # Direct attribute access

    print('Prepared mesh.')


    geodesic = trimesh.load(geodesic_filepath)

    
    edge_id = None

    vtx_id = first_vtx_id

    print('Splitting edges.')
    
    for i in range(1, geodesic.vertices.shape[0] - 1):
        
        print(i, 'out of', geodesic.vertices.shape[0]-1)



            
        p = geodesic.vertices[i,:]
        onering_halfedges = ordered_onering_halfedges(he_mesh, V_he, vtx_id)
        #print('values are:')

        best_dist = 1000

        A=None
        B=None
        
        for j in onering_halfedges:
            #print('he:',j)

            
            
            a = he_mesh.vertices[he_mesh.half_edges[j].vertex_indices[0]]
            b = he_mesh.vertices[he_mesh.half_edges[j].vertex_indices[1]]
            
            if debug_plot:
                plt.plot([a[0], b[0] ], [a[1], b[1]], label=j)
                plt.scatter(p[0], p[1], color='magenta')
                plt.scatter(he_mesh.vertices[vtx_id][0], he_mesh.vertices[vtx_id][1], color='blue')

            cur_dist = dist(p,a,b)
            if cur_dist<best_dist and cur_dist<0.0001 :
                print('Point p is on halfedge', j)
                edge_id = j
                best_dist = cur_dist
                A=a
                B=b
                
            #if point_on_segment(p, a, b):
            #    print('Point p is on halfedge', j)
            #    
            #    edge_id = j
            #    A=a
            #    B=b

            
        if debug_plot:        
            plt.legend()
            plt.show()
            
        skip=False
        if ((p-A)**2).sum()<0.00001:
            skip=True
            vtx_id = he_mesh.half_edges[edge_id].vertex_indices[0]
            
        if ((p-B)**2).sum()<0.00001:
            skip=True
            vtx_id = he_mesh.half_edges[edge_id].vertex_indices[1]
            
            


        
            
        if not (edge_id is None) and not skip:
            print('next edge to split:',edge_id)
    
            he_mesh, V_he = edge_split(he_mesh, V_he, edge_id=edge_id, vtx_position=p)
            vtx_id = len(he_mesh.vertices) - 1 #it's the newest vertex to be added
            
        elif not skip:
            print('the point is not on a line segment...')
            break
            
        #print(he_mesh)
        #o3d.visualization.draw_geometries([he_mesh], window_name="Egde Split Mesh")
            

        ### very inefficient!!!
        #V_he = [0 for _ in range(len(he_mesh.vertices))]
        #for k in range(len(he_mesh.half_edges)):
        #    V_he[he_mesh.half_edges[k].vertex_indices[0]] = k
        
        edge_id=None

        

    o3d.visualization.draw_geometries([he_mesh], window_name="Final Refined Mesh")
    


    #he_mesh = edge_split(he_mesh, edge_id=1, test_noise=True)
    #he_mesh = edge_split(he_mesh, edge_id=2, test_noise=True)
    #he_mesh = edge_split(he_mesh, edge_id=6, test_noise=True)
    
    
    

    mesh_out = trimesh.Trimesh(vertices=he_mesh.vertices, faces=he_mesh.triangles)

    # Save as OBJ file
    mesh_out.export("output_mesh.obj")

```

```python

first_vtx_id = 3
last_vtx_id = 1

cut_mesh_along_geodesic('../geodesic_matlab/data/sphere1.off',
                        '../geodesic_matlab/paths/sphere1_1_3_path.obj',
                        first_vtx_id, last_vtx_id, debug_plot=True)
```
