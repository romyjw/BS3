```python
# define the positional encoding to use, and consistent mlp layer sizes.

n_posenc_levels = 0
#hidden_layer_sizes = [64,64,64,64]

#hidden_layer_sizes = [64,64]
#hidden_layer_sizes = [128, 128]
hidden_layer_sizes = [32]
#hidden_layer_sizes = [16]

hidden_layer_sizes = [128]

hidden_layer_sizes = [4]

hidden_layer_sizes = [128,128]


#hidden_layer_sizes = [128, 128, 128, 128]

#hidden_layer_sizes = [256,256,256,256]


blend_type = 'pou_2' #'bary' #'pou_2' #'discrete' #'2d_pou' #'radial' #'bary'

# which shape and which patches will the BNS train on?
shapename = '100face_torus' #'sphere2_20patch' #'400face_torus'#'igea11706_6patch'   #'100face_torus'
angle_flag = 'equal' # equal


# 0.15470053837 < v < 0.73205080756
overlap_param = 0.7



```

```python


def posenc(x):#x is (N, 2)
    
    if n_posenc_levels == 0:
        return x
        
    p = 0.2 * x  #multiply by 0.2 so that no cyclic mess-up.  
    freqs = 2 ** torch.arange(0, n_posenc_levels, device=x.device).float()  # [1, 2, 4, 8, 16, 32, 64, 128]
    p = p.unsqueeze(-1) * freqs * two_pi  # (N, 2, 8)

    encoded = torch.cat([torch.sin(p), torch.cos(p)], dim=-1)  # (N, 2, 16)
    return encoded.view(x.shape[0], -1)  # Flatten to (N, 32)

if n_posenc_levels == 0:
    layer_sizes = [2] + hidden_layer_sizes + [3]
else:
    layer_sizes = [n_posenc_levels * 4] + hidden_layer_sizes + [3]


import BNS
importlib.reload(BNS)
from BNS import BNS


# instantiate
bns = BNS(layer_sizes, from_file=True, shapename= shapename, posenc=posenc, angle_flag=angle_flag, overlap_param = overlap_param, virtual_vertex_flag=False, blend_type=blend_type)
total_params = sum(p.numel() for p in bns.parameters())
print('Number of Parameters:', total_params)


select_patch_indices=None
#select_patch_indices = bns.onerings[0]['triangles']


```

```python
############### initial training settings #####################

epoch=0

losses=[]
pos_losses=[]
normals_regs=[]
fractional_epochs=[]

epochs=[]
times=[]
num_patches = bns.num_patches

loss=99999

optimizer = optim.Adam(bns.polynomial_parameters, lr=0.001)
#optimizer = optim.SGD(bns.parameters(), lr=0.001)
```

```python
##################### define training loop ###########################

import torch
import time
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display, clear_output



def train_batch_loop(
    lr=0.01,
    batch_size=10000,
    min_loss=0.0005,
    normals_reg_coeff=0.0,
    batch_print_rate=100,
    batch_plot_rate=10,
    epoch_print_rate=1,
    epoch_save_rate=1,
    epoch_plot_rate=1,
    plot=False,
    flag='torus_energy_poly'#'unitnorm_energy_poly' #'poly' #'unblended_poly'
):
    global optimizer, losses, pos_losses, normals_regs, times, epoch, loss, num_patches

    

    for g in optimizer.param_groups:
        g['lr'] = lr

    seqs = t_train.shape[0]
    total_points = t_train.shape[1]

    print(t_train.shape)

    updated_target_train = target_train
    print(updated_target_train.shape)

    while loss > min_loss:
        perm = torch.randperm(total_points)
        epoch_loss = 0.0

        for i in range(0, total_points, batch_size):
            batch_start = time.time()

            idx = perm[i:i + batch_size]
            batch_t = t_train[:, idx, :]
            batch_target = updated_target_train[:, idx, :]
            
            batch_target_normals = target_normals[:, idx, :]

            optimizer.zero_grad()

            batch_t.requires_grad = True
            output_train = bns(batch_t, select_patch_indices=select_patch_indices, test_flag=flag)

            if normals_reg_coeff > 0.0:
                output_normals = diffmod.compute_normals(out=output_train, wrt=batch_t)
                normals_reg = normals_reg_coeff * (output_normals - batch_target_normals).pow(2).sum(-1).mean()
            else:
                normals_reg = torch.tensor(0.0)

            if flag == 'unblended_poly':
                #print('yo')
                #print (output_train.requires_grad)

                #print('out, targ', output_train.shape, batch_target.shape)
                position_loss_list = []
                
                for j in range(bns.V.shape[0]):
                    cur_face_indices = bns.onerings[j]['triangles']
                    diff = output_train[cur_face_indices, j , :,:] - batch_target[cur_face_indices,:,:]
                    loss_j = diff.pow(2).sum(-1).mean()
                    position_loss_list.append(loss_j)
                
                position_loss = torch.stack(position_loss_list).mean()
                
            elif flag=='poly':
                position_loss = (output_train[select_patch_indices,:,:] - batch_target[select_patch_indices,:,:]).pow(2).sum(-1).mean()
                
            elif flag=='unitnorm_energy_poly':
                position_loss = ( output_train[select_patch_indices,:,:].pow(2).sum(-1) -1.0 ).pow(2).mean()

            elif flag=='torus_energy_poly':
                print('ot', output_train.shape)
                
                theta = torch.atan2(output_train[select_patch_indices,:,2], output_train[select_patch_indices,:,0])
                central_ring = 0.32 * torch.stack([torch.cos(theta), 0*torch.cos(theta), torch.sin(theta)])
                print(central_ring.shape, output_train[select_patch_indices,:,:].shape)

                
                position_loss = ( (output_train[select_patch_indices,:,:] - central_ring).pow(2).sum(-1) - 0.1**2 ).pow(2).mean()
                
            
            batch_loss = position_loss + normals_reg
            batch_loss.backward()
            optimizer.step()

            epoch_loss += batch_loss * batch_t.shape[1]

            batch_time = time.time() - batch_start

            times.append(batch_time)
            losses.append(batch_loss.detach().item())
            pos_losses.append(position_loss.detach().item())
            normals_regs.append(normals_reg.detach().item())
            fractional_epochs.append(epoch + i/total_points)

            

            if i % batch_plot_rate == 0 and plot:
                
                fig, axs = plt.subplots(2, 1, figsize=(10, 6))
            
                # Plot loss curves
                axs[0].plot(fractional_epochs, np.log10(losses),  label='loss')
                axs[0].plot(fractional_epochs, np.log10(pos_losses),  label='pos loss')
                if normals_reg_coeff > 0:
                    axs[0].plot(fractional_epochs, np.log10(normals_regs),  label='normals reg')
                axs[0].legend()
                axs[0].set_xlabel('Epoch')
                axs[0].set_ylabel('Log10 of Loss')
                axs[0].set_title('Log10 of Losses')
            
                # Add latest batch loss as text on plot
                axs[0].text(
                    0.8, 0.9,
                    f"Batch Loss: {batch_loss.item():.8f}",
                    transform=axs[0].transAxes,
                    ha='right', va='bottom',
                    fontsize=10, color='black',
                    bbox=dict(facecolor='white', alpha=0.8, edgecolor='gray')
                )
                axs[0].text(
                    0.8, 0.8,
                    f"Batch Pos. Loss: {position_loss.item():.8f}",
                    transform=axs[0].transAxes,
                    ha='right', va='bottom',
                    fontsize=10, color='black',
                    bbox=dict(facecolor='white', alpha=0.8, edgecolor='gray')
                )

                if normals_reg_coeff > 0:
                    axs[0].text(
                        0.8, 0.7,
                        f"Batch Normals Reg.: {normals_reg.item():.8f}",
                        transform=axs[0].transAxes,
                        ha='right', va='bottom',
                        fontsize=10, color='black',
                        bbox=dict(facecolor='white', alpha=0.8, edgecolor='gray')
                    )
            
                # Plot iteration time
                axs[1].plot( fractional_epochs, times)
                axs[1].set_xlabel('Epoch')
                axs[1].set_title('Time per Batch')
            
                # Add latest batch time as text
                axs[1].text(
                    0.99, 0.90,
                    f"Batch Time: {batch_time:.4f}s",
                    transform=axs[1].transAxes,
                    ha='right', va='bottom',
                    fontsize=10, color='black',
                    bbox=dict(facecolor='white', alpha=0.8, edgecolor='gray')
                )
            
                plt.tight_layout()
                display(fig)
                plt.close(fig)




            #if i%100000 == 0 and i>10:
            #    print('Re-computing ground truth correspondence')
            #    print(updated_target_train.shape)
            #    updated_target_train, normals = project_to_torus(updated_target_train)

            

            if normals_reg_coeff > 0:
                print(f"Batch {i}, Loss: {batch_loss:.8f}, Pos. Loss: {position_loss:.8f}, "
                          f"Normals Reg: {(normals_reg_coeff * normals_reg):.8f}, Batch Time: {batch_time:.4f}s")
            else:
                print(f"Batch {i}, Loss: {batch_loss:.8f}, Pos. Loss: {position_loss:.8f}," f"Batch Time: {batch_time:.4f}s")

            total_batch_time = time.time() - batch_start

            print(f"Time inc plotting: {total_batch_time: .4f}s")
                
            clear_output(wait=True)
            
                

        loss = torch.tensor(epoch_loss / total_points)

        if epoch % epoch_print_rate == 0:
            print(f"Epoch {epoch}, Loss: {loss.item():.8f}")

        if epoch % epoch_save_rate == 0:
            bns.save_mlps('models/temp.pth')
            print('Saved checkpoint.')






        epoch += 1
```
