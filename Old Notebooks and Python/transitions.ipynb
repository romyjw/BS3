```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact
from matplotlib.patches import Circle

def line_from_points(p1, p2):
    """Returns the slope and intercept of the line passing through two points."""
    if p1[0] == p2[0]:
        return None, p1[0]  # Vertical line
    m = (p2[1] - p1[1]) / (p2[0] - p1[0])
    b = p1[1] - m * p1[0]
    return m, b

def perpendicular_bisector(p1, p2):
    """Finds the perpendicular bisector of a segment defined by two points."""
    mid = ((p1[0] + p2[0]) / 2, (p1[1] + p2[1]) / 2)
    m, _ = line_from_points(p1, p2)
    
    if m is None:
        return 0, mid[1]  # Horizontal line
    if m == 0:
        return None, mid[0]  # Vertical line
    
    m_perp = -1 / m
    b_perp = mid[1] - m_perp * mid[0]
    return m_perp, b_perp

def intersection(m1, b1, m2, b2):
    """Finds intersection of two lines given their slopes and intercepts."""
    if m1 is None:
        return (b1, m2 * b1 + b2)
    if m2 is None:
        return (b2, m1 * b2 + b1)
    if m1 == m2:
        return None
    x = (b2 - b1) / (m1 - m2)
    y = m1 * x + b1
    return x, y

def plot_triangle(x1, y1, x2, y2, x3, y3):
    """Plots the triangle, perpendicular bisectors, and orthocenter dynamically."""
    points = np.array([[x1, y1], [x2, y2], [x3, y3]])
    
    fig, ax = plt.subplots(figsize=(6, 6))
    ax.set_xlim(0, 10)
    ax.set_ylim(0, 10)

    # Draw triangle
    for i in range(3):
        p1, p2 = points[i], points[(i+1) % 3]
        ax.plot([p1[0], p2[0]], [p1[1], p2[1]], 'b')

    # Perpendicular bisectors
    bisectors = []
    for i in range(3):
        p1, p2 = points[i], points[(i+1) % 3]
        m, b = perpendicular_bisector(p1, p2)
        if m is not None:
            x_vals = np.linspace(0, 10, 100)
            y_vals = m * x_vals + b
            ax.plot(x_vals, y_vals, 'g--')
        else:
            ax.axvline(x=b, color='g', linestyle='--')
        bisectors.append((m, b))

    # Orthocenter
    orthocenter = intersection(bisectors[0][0], bisectors[0][1], bisectors[1][0], bisectors[1][1])
    if orthocenter:
        ax.scatter(*orthocenter, color='purple', s=100, label="Orthocenter")

    # Draw triangle vertices
    for (x, y) in points:
        ax.add_patch(Circle((x, y), 0.2, color='red'))
    
    ax.legend()
    plt.show()

# Interactive widget
interact(plot_triangle, 
         x1=(1, 9, 0.1), y1=(1, 9, 0.1), 
         x2=(1, 9, 0.1), y2=(1, 9, 0.1), 
         x3=(1, 9, 0.1), y3=(1, 9, 0.1));
```
