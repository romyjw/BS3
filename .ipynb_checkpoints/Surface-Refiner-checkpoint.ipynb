```python
import torch
import torch.nn as nn
import torch.optim as optim

# Define a simple MLP with customizable layer sizes
class MLP(nn.Module):
    def __init__(self, layer_sizes):
        super(MLP, self).__init__()
        layers = []
        for i in range(len(layer_sizes) - 1):
            layers.append(nn.Linear(layer_sizes[i], layer_sizes[i + 1]))
            if i < len(layer_sizes) - 2:
                layers.append(nn.Tanh())  # activation between hidden layers
        self.model = nn.Sequential(*layers)

    def forward(self, x):
        return self.model(x)




```

```python
def disp_modifier_orthog(bs_points, bs_normals, model, scale=0.001):
    disp = model(bs_points)
    return bs_points + disp*scale*bs_normals

def disp_modifier_general(bs_points, bs_normals, model, scale=10.0):
    disp = model(bs_points)
    return bs_points + disp*scale
```

```python
layer_sizes = [3, 20, 3]
disp_mlp = MLP(layer_sizes)

# Dummy optimizer
optimizer = optim.Adam(disp_mlp.parameters(), lr=0.01)





```

```python
# Training loop
for epoch in range(10000):




    training_sample_dict = bns.compute_samples(num_samples=1000, avoid_corners=True)
    x = training_sample_dict['uv']
    x.requires_grad=True
    
    bns.disp_modifier = None
    bs_points = bns(x, test_flag='poly')
    #bs_normals = diffmod.compute_normals(out=bs_points, wrt=x)
    
    
    bs_points = bs_points.detach().clone().requires_grad_(True)
    #bs_normals = bs_normals.detach().clone().requires_grad_(True)




    
    
    optimizer.zero_grad()

    new_points = disp_modifier_general(bs_points, None,  model=disp_mlp)

    

    # --- Placeholder for loss function ---
    # Replace this with your actual loss, e.g. nn.MSELoss()(output, y)
    loss = sdf( new_points , shape_id, squared=True).mean()

    if epoch==0:
        baseline_loss = sdf( bs_points , shape_id, squared=True).mean()
        print(f"Epoch {epoch}: baseline loss = {baseline_loss.item():.10f}")


    loss.backward()
    optimizer.step()
    if epoch%10==0:
        print(f"Epoch {epoch}: loss = {loss.item():.10f}")
```

```python
bns.disp_modifier = lambda bs_points, bs_normals : disp_modifier_general(bs_points, bs_normals, model=disp_mlp)
#bns.disp_modifier = None
```
